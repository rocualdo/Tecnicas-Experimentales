# Regresiones lineales

Este Notebook contiene un módulo con el código necesario para realizar regresiones lineales simples y ponderadas, además de dos programas para poder llevar a cabo cada una de ellas llamando a las funciones definidas en dicho módulo.

```python
#MÓDULO CON LAS FUNCIONES DE LOS DOS TIPOS DE REGRESIONES

from numpy import*
def rls(x,y):
    if len(x)!=len(y):
        print("Vectores de dimensions distintas")
    else:
        n=len(x)
        p=dot(x,y)
        sx=sum(x)
        sy=sum(y)
        x2=x**2
        sx2=sum(x2)
        y2=y**2
        sy2=sum(y2)
        b=((n*p)-(sx*sy))/((n*sx2)-(sx**2))
        a=((sx2*sy)-(sx*p))/((n*sx2)-(sx**2))
        c=0
        for i in range(len(x)):
            c=c+((y[i]-a-b*(x[i]))**2)
        s=sqrt((c)/(n-2))
        sa=s*(sqrt((sx2)/((n*sx2)-(sx**2))))
        sb=s*(sqrt(n/((n*sx2)-(sx**2))))
        r=((n*p)-(sx*sy))/(sqrt(((n*sx2)-(sx**2))*((n*sy2)-(sy**2))))
        lista=list([a,sa,b,sb,s,r])
    return lista
def rlp(x,y,sy):
    if len(x)!=len(y):
        print("Vectores de dimensions distintas")
    elif len(y)!=len(sy):
        print("Vectores de dimensions distintas")
    else:
        n=len(x)
        w=[0.0]*n
        for i in range(n):
            w[i]=1/((sy[i])**2)
        w=array(w)
        x2=x**2
        y2=y**2
        wxy=0
        for i in range(len(x)):
            wxy=wxy+((w[i])*(x[i])*(y[i]))
        sw=sum(w);wx=dot(w,x);wx2=dot(w,x2);wy=dot(w,y);wy2=dot(w,y2)
        det=((sw)*(wx2))-((wx)**2)
        a=((wy*wx2)-(wx*wxy))/det
        b=((sw*wxy)-(wx*wy))/det
        sa=sqrt(wx2/det)
        sb=sqrt(sw/det)
        c=0
        for i in range(len(x)):
            c=c+((w[i])*(((y[i])-a-(b*(x[i])))**2))
        f=n/((n-2)*sw)
        s=sqrt(f*c)
        r=((sw*wxy)-(wx*wy))/(sqrt(((sw*wx2)-((wx)**2))*((sw*wy2)-(wy**2))))
        lista=list([a,sa,b,sb,s,r])
    return lista
```

```python
#REGRESION LINEAL SIMPLE

from numpy import*
from regresiones import rls
print("REGRESION LINEAL SIMPLE")
print("Introduce valores de x e y:")
x=input("Introduce a secuencia de numeros (espaciados) do vector x: ").split()
v=float_(x) #transformamos esa lista de cadeas de carct. nun array numerico
y=input("Introduce a secuencia de numeros (espaciados) do vector y: ").split()
w=float_(y)
resul=rls(v,w)
print("..................................")
print("Resultados da regresion:")
print("a:",resul[0])
print("s(a):",resul[1])
print("b:",resul[2])
print("s(b):",resul[3])
print("s:",resul[4])
print("r:",resul[5])
print("..................................")
```
