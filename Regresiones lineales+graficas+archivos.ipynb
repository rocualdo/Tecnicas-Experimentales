# Regresiones lineales, gráficas y archivos

El programa siguiente permite realizar regresiones lineales simples o ponderadas (llamando a las funciones definidas en el modulo de regresiones disponible en este repositorio), dependiendo de lo que se necesita. Además de calcular los parámetros necesarios después de introducir los valores de x e y experimentales, permite guardarlos ordenadamente en un archivo (de tipo y nombre a escoger) junto con cualquier nota o comentario.

Las distintas opciones se podrán llevar a cabo contestando con 1 (sí) o 2 (no) cuando se indique. Es necesario cerrar la gráfica, que se mostrará en pantalla al ejecutar el programa desde la terminal, para poder continuar y guardar un archivo con los resultados de las regresiones.

_Nota: asegúrate de tener el módulo de regresiones lineales!_

```python
from pylab import*
from numpy import*
from regresiones import*
xx=input("Introduce valores de las x equiespaciados: ").split()
yy=input("Introduce valores de las y equiespaciados: ").split()
x=float_(xx)
y=float_(yy)
a=int(input("Indica tipo de regresion lineal, simple (1) o ponderada (2): "))
if a==1:
    lista1=rls(x,y)
    pen1=lista1[2]
    b1=lista1[0]
    print("Parametros de la regresion lineal (Recta y=bx+a):")
    print("a=",lista1[0])
    print("s(a)=",lista1[1])
    print("b=",lista1[2])
    print("s(b)=",lista1[3])
    print("s=",lista1[4])
    print("r=",lista1[5])
if a==2:
    syy=input("Introduce valores de s(y) equiespaciados: ").split()
    sy=float_(syy)
    lista2=rlp(x,y,sy)
    pen=lista2[2]
    b=lista2[0]
    print("Parametros de la regresion lineal (Recta y=bx+a):")
    print("a=",lista2[0])
    print("s(a)=",lista2[1])
    print("b=",lista2[2])
    print("s(b)=",lista2[3])
    print("s=",lista2[4])
    print("r=",lista2[5])

ejex=linspace(min(x),max(x),100)
if a==1:
    plot(x,y,linewidth=0.0,marker="+",color="red")
    ejey=b1+pen1*ejex
    plot(ejex,ejey,linewidth=1.5,color="blue")
if a==2:
    plot(x,y,linewidth=0.0,marker="+",color="red")
    ejey=b+pen*ejex
    plot(ejex,ejey,linewidth=1.5,color="blue")
    
xlabel("x")
ylabel("y")
show()

q=int(input("¿Quieres guardar los resultados de la regresion en un archivo? Si(1) No(2) "))
if q==1:
    nome=input("Introduce nombre del archivo (pon .txt): ")
    outfile=open(nome,"w")
    nota=input("Introduce aclaracion del contenido: ")
    nota=nota+"\n"
    outfile.write(nota)
    letras=(["a:   ","s(a):   ","b:   ","s(b):   ","s:   ","r:   "])
    if a==1:
        for i in range(len(lista1)):
            aux1=""
            aux1=aux1+letras[i]+str(lista1[i])
            aux1=aux1+"\n"
            outfile.write(aux1)
        aux=input("Comentarios: ")
        outfile.write(aux)
        outfile.close()
    if a==2:
        for i in range(len(lista2)):
            aux2=""
            aux2=aux2+letras[i]+str(lista2[i])
            aux2=aux2+"\n"
            outfile.write(aux2)
        aux=input("Comentarios: ")
        outfile.write(aux)
        outfile.close()
```
